#Central limit theorem
Distribution of sample means approximate Gaussian Normal

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#how to create samples
np.random.uniform(size = 100)
```

```python
#compute the mean of the created numbers
np.mean(np.random.uniform(size = 100))
```

```python
#create a mean with 10000 samples
sample = lambda: np.mean(np.random.uniform(0,1, size = 100))
x = [sample() for _ in range(10000)]
```

```python
#plot aa histogram of the 10,000 sample means
plt.hist(x, bins = 40, edgecolor = 'black')
```

#Experiment with Dice Throwing

```python
x = lambda n:np.mean(np.random.randint(1, 7, size = n))
```

```python
k = 1
for n in [1, 2, 5, 10]:
  xx = [x(n) for _ in range(10000)]
  plt.subplot(2, 2, k)
  plt.hist(xx, bins = 40, density = True, edgecolor = 'black')
  k = k + 1
plt.show()
```
